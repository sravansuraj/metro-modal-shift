```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="darkgrid")
os.makedirs("../outputs", exist_ok=True)
print("✅ All imports successful!")
```

```python
df = pd.read_csv("../data/india_transport_modal_shift.csv")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
print("Dataset Info:")
print(df.info())
print("\nBasic Statistics:")
df.describe()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Year'], df['Private_Vehicles_Registered_Millions'], marker='o', color='tomato', linewidth=2)
plt.title('Private Vehicle Registrations in India (2001–2023)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Vehicles Registered (Millions)')
plt.xticks(df['Year'], rotation=45)
plt.tight_layout()
plt.savefig('../outputs/plot1_private_vehicles.png', dpi=150)
plt.show()
print("✅ Plot 1 saved!")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Year'], df['Public_Transport_Share_Percent'], marker='s', color='steelblue', linewidth=2)
plt.title('Public Transport Modal Share % in India (2001–2023)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Public Transport Share (%)')
plt.xticks(df['Year'], rotation=45)
plt.tight_layout()
plt.savefig('../outputs/plot2_public_transport_share.png', dpi=150)
plt.show()
print("✅ Plot 2 saved!")
```

```python
plt.figure(figsize=(12, 5))
plt.bar(df['Year'], df['Metro_Ridership_Million_Trips'], color='mediumseagreen')
plt.title('Metro Ridership in India (2001–2023)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Ridership (Million Trips)')
plt.xticks(df['Year'], rotation=45)
plt.tight_layout()
plt.savefig('../outputs/plot3_metro_ridership.png', dpi=150)
plt.show()
print("✅ Plot 3 saved!")
```

```python
fig, ax1 = plt.subplots(figsize=(13, 6))

ax1.set_xlabel('Year')
ax1.set_ylabel('Private Vehicles (Millions)', color='tomato')
ax1.plot(df['Year'], df['Private_Vehicles_Registered_Millions'], color='tomato', marker='o', linewidth=2, label='Private Vehicles')
ax1.tick_params(axis='y', labelcolor='tomato')

ax2 = ax1.twinx()
ax2.set_ylabel('Public Transport Share (%)', color='steelblue')
ax2.plot(df['Year'], df['Public_Transport_Share_Percent'], color='steelblue', marker='s', linewidth=2, linestyle='--', label='Public Transport Share')
ax2.tick_params(axis='y', labelcolor='steelblue')

plt.title('Private Vehicle Growth vs Public Transport Share (2001–2023)', fontsize=14)
fig.tight_layout()
plt.savefig('../outputs/plot4_private_vs_public.png', dpi=150)
plt.show()
print("✅ Plot 4 saved!")
```

```python
plt.figure(figsize=(9, 6))
sns.heatmap(df.drop('Year', axis=1).corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Transport Variables', fontsize=14)
plt.tight_layout()
plt.savefig('../outputs/plot5_correlation_heatmap.png', dpi=150)
plt.show()
print("✅ Plot 5 saved!")
```
